### 
用for迴圈抓取65個產品品項的名稱(name)、價錢(price)及評分(rating)，
有一些品項沒有reviews，因此要寫if，若該商品沒有reviews，則設定其rating為none，
若該商品有reviews，則顯示其評分

```python
products=[]
for i in range(65):
    if res.json()['products'][i]['reviews']==None:
        df=pd.DataFrame([{'product name':res.json()['products'][i]['name'],
                        'price':res.json()['products'][i]["pricing"]['current']['price'],
                        'rating':'None'
                      }] )
        products.append(df)
    else:
        df=pd.DataFrame([{'product name':res.json()['products'][i]['name'],
                          'price':res.json()['products'][i]["pricing"]['current']['price'],            
                          'rating':res.json()['products'][i]['reviews']["rating"]
                     }] )
        products.append(df)   
```

```python
products_all = pd.concat(products,ignore_index=True)
```

```python
products_all
```
