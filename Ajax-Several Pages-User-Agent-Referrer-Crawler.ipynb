```python
import requests
import pandas as pd
import numpy as np 
```

```python
def start_requests(url,hs):    
    res = requests.get(url,headers=hs)
    return res.json()
```

```python
def parse(json):
    global df
    df=pd.DataFrame()
    products_info=[]
    for i in range(60):
        if json['products'][i]['reviews']==None:
            df=pd.DataFrame([{'product name':json['products'][i]['name'],
                              'price':json['products'][i]["pricing"]['current']['price'],
                              'rating':'None'
                             }] )
            products_info.append(df)
        else:
            df=pd.DataFrame([{'product name':json['products'][i]['name'],
                              'price':json['products'][i]["pricing"]['current']['price'],            
                              'rating':json['products'][i]['reviews']["rating"]
                             }] )
            products_info.append(df)
    return  products_info
```

```python
def Get_all_page(j):
    All_products=pd.DataFrame()
    temp=pd.DataFrame()
    for i in range(1,j):
        url = 'https://www.overstock.com/Home-Garden/Living-Room-Furniture/713/cat.html?format=fusion&client_id=sn_spa&count=60&page={}'.format(i)
        hss = {
            "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36",
            "referrer":'https://www.overstock.com/Home-Garden/Living-Room-Furniture/713/cat.html?page={}'.format(i)
        }
        json=start_requests(url,hss)
        parse(json)
        temp=pd.concat(parse(json),ignore_index=True)
        All_products=pd.concat([All_products,temp],ignore_index=True)     
    return All_products
```
