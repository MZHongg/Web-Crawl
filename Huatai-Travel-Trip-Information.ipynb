```python
from pyquery import PyQuery
```

```python
HW = PyQuery(res_HW.text)
```

```python
PyQuery(HW(".thumbnail")[1]).outerHtml()
```

```python
HW(".thumbnail").map(lambda i,e:{
     "旅遊天數": PyQuery(e)(".product_days").text()
    ,"行程名稱": PyQuery(e)(".product_name").text()
    ,"出發日期": PyQuery(e)(".product_date").text()
    ,"可售位、團位": PyQuery(e)(".product_available").text()
    ,"價錢": PyQuery(e)(".product_price").text()
})
```

```python
import pandas as pd
```

```python
HW_Data = pd.DataFrame(HW(".thumbnail").map(lambda i,e:{
    "旅遊天數": PyQuery(e)(".product_days").text()
    ,"行程名稱": PyQuery(e)(".product_name").text()
    ,"出發日期": PyQuery(e)(".product_date").text()
    ,"可售位、團位": PyQuery(e)(".product_available").text()
    ,"價錢": PyQuery(e)(".product_price").text()
}))

```
